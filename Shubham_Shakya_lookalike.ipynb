```python
import pandas as pd
```

```python
c = pd.read_csv("Customers.csv")
p = pd.read_csv("Products.csv")
t = pd.read_csv("Transactions.csv")
```

```python
m = t.merge(c, on="CustomerID", how="left")
```

```python
m= m.merge(p, on="ProductID", how="left")
```

```python
m
```

```python
m = m.drop('Price_x',axis=1)
```

```python
m.rename(columns = {"Price_y" : "Price"},inplace = True)
```

```python
m
```

```python
customer_features = (
    m.groupby("CustomerID")
    .agg(
        avg_transaction_value=("TotalValue", "mean"),
        total_purchases=("Quantity", "sum"),
        total_spent=("TotalValue", "sum"),
        num_transactions=("TransactionID", "count"),
    )
    .reset_index()
)
```

```python
c_profiles = pd.merge(c, customer_features, on="CustomerID", how="left")
c_profiles = pd.get_dummies(c_profiles, columns=["Region"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Normalize numeric features for similarity calculation
numeric_features = c_profiles.select_dtypes(include=[np.number])
normalized_features = (numeric_features - numeric_features.min()) / (numeric_features.max() - numeric_features.min())

# Compute pairwise cosine similarity
similarity_matrix = cosine_similarity(normalized_features)
```

```python
lookalike_results = {}
for i in range(20):  # First 20 customers
    customer_id = c_profiles.iloc[i]["CustomerID"]
    similarity_scores = similarity_matrix[i]
    top_indices = np.argsort(similarity_scores)[::-1][1:4]  # Exclude the customer itself
    top_customers = [
        (c_profiles.iloc[idx]["CustomerID"], similarity_scores[idx])
        for idx in top_indices
    ]
    lookalike_results[customer_id] = top_customers

```

```python
# Convert to DataFrame for Lookalike.csv
lookalike_df = pd.DataFrame(
    [
        {"CustomerID": cust, "Lookalikes": str([(c[0], round(c[1], 2)) for c in lookalikes])}
        for cust, lookalikes in lookalike_results.items()
    ]
)
```

```python
lookalike_df.to_csv("Shubham_Shakya_Lookalike.csv", index=False)
```
